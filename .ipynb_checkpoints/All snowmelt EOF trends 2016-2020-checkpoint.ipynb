```python
from pathlib import Path
import numpy as np
import datetime
import netCDF4
import matplotlib.pyplot as plt
import pyproj
import cartopy.feature as cfe
import cartopy.crs as ccrs
import pandas as pd
from pyproj import Transformer
from pyproj import CRS
from eofs.xarray import Eof
from shapely.geometry.polygon import Polygon
import shapefile as shp
import fiona 
import rioxarray
import geopandas as gpd
from shapely.geometry import mapping
import geoplot
from rasterio import features
from shapely.geometry import Point
import matplotlib.colors as colors
from numpy import ma
import xarray as xr
```

```python
filename="snowmelt.KNMI-2016.XPEN055.ERA5-3H_RACMO2.3p2.3H.nc"
this_dir=Path.cwd()
data_dir=this_dir.parent/Path("Data")
files = data_dir.rglob(filename)
for i in files:
    data = i
```

```python
snowmelt_data  = xr.open_dataset(data)
snowmelt_data
```

## Decimal Coordinates of boundary and AWS locations

```python
## Points to bound GVIIS area:
ul_lat= -68.8057057
ul_lon= -70.4385094
ur_lat= -68.9444216
ur_lon= -65.1641375
ll_lat= -73.5935193
ll_lon= -78.5115361
lr_lat= -74.0007447
lr_lon= -67.0772934


#SPRI AWS STATION COORDINATE
AWSlon = -67.69971
AWSlat = -71.14409

#BAS FOSSIL BLUFF AWS STATION COORDINATE
BAS_FB_lon = -68
BAS_FB_lat = -71.329
```

```python
data_crs=CRS('-m 57.295779506 +proj=ob_tran +o_proj=latlon +o_lat_p=-180.0 +lon_0=30.0')
from_lat_lon_crs = CRS.from_proj4("+proj=latlon")
transproj = Transformer.from_crs(from_lat_lon_crs, data_crs)
```

## Rotate the points

```python
rotateds= transproj.transform((ll_lon,ur_lon), (ul_lat,lr_lat))
print('data slicing boundary = ',rotateds)
lats_r=rotateds[1]
lons_r=rotateds[0]

##Boundary points rotated
rotated_pts=transproj.transform((ul_lon,ur_lon,ll_lon,lr_lon), (ul_lat,ur_lat,ll_lat,lr_lat))
gviis_lons=rotated_pts[0]
print('GVIIS boundary lons =',gviis_lons)
gviis_lats=rotated_pts[1]
print('GVIIS boundary lats =',gviis_lats)

##SPRI AWS STATION ROTATED
aws_rotated=transproj.transform(AWSlon,AWSlat)
print('AWS rotated = ',aws_rotated)


##BAS FOSSIL BLUFF AWS STATION ROTATED
bas_fb_rotated=transproj.transform(BAS_FB_lon,BAS_FB_lat)
print('BAS Fossil Bluff rotated = ',bas_fb_rotated)
```

# Plot Data with markers for rotated GVIIS boundary and AWS stations

```python
melt=snowmelt_data.snowmelt.squeeze()
data2= melt.isel(time=0)
rlon = snowmelt_data.rlon.values
rlat = snowmelt_data.rlat.values


plt.figure(figsize=(10,10))

ax = plt.subplot( projection=ccrs.Stereographic(central_longitude=-67.69971, central_latitude=-71.14409))
ax.set_extent([-79,-65,-75,-66])

data_crs = ccrs.RotatedPole(pole_longitude=snowmelt_data.rotated_pole.grid_north_pole_longitude,
                            pole_latitude=snowmelt_data.rotated_pole.grid_north_pole_latitude)

# plot data
result = ax.contourf(rlon, rlat, data2, extend='both', cmap='Blues', transform=data_crs)
poly_coords = Polygon(((gviis_lons[0], gviis_lats[0]),
                       (gviis_lons[1], gviis_lats[1]),
                       (gviis_lons[3], gviis_lats[3]),
                       (gviis_lons[2], gviis_lats[2])))


# ax.add_patch(Polygon(poly_coords, color='forestgreen', alpha=0.5))
ax.plot(gviis_lons[0], gviis_lats[0], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[1], gviis_lats[1], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[2], gviis_lats[2], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[3], gviis_lats[3], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(aws_rotated[0], aws_rotated[1], markersize=7, marker='^', color='green',transform=data_crs,label='SPRI AWS')
ax.plot(bas_fb_rotated[0], bas_fb_rotated[1], markersize=7, marker='^', color='orange',transform=data_crs,label='BAS AWS')

plt.legend()

ax.coastlines(resolution='10m')
ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '10m', edgecolor='k'))
plt.title("Snowmelt Flux on the GVIIS")
plt.colorbar(result, orientation='vertical', label='kg $m^{-2}$ $s^{-1}$', extend='both', fraction=0.046, pad=0.04)
ax.add_geometries([poly_coords], crs=data_crs, facecolor="none",edgecolor='red', alpha=0.8,linewidth=3)

plt.show()
```

# Create functions to:
# 1. Handle resamples of desired frequency and return plotting lons and lats
# 2. Filter summer months only
# 3. Plot EOF Maps
# 4. Plot EOF PC Amplitudes

```python
def resample_bounds(data,frequency):
    d2=data.resample(time=frequency).mean()
    lat = d2['rlat'].values
    lon = d2['rlon'].values
    melts = d2['snowmelt'].values
    lons, lats = np.meshgrid(lon, lat)
    return d2,lons,lats

def resample_summer(data,summer_months):
    d2 = data.sel(time=data.time.dt.month.isin(summer_months))
    lat = d2['rlat'].values
    lon = d2['rlon'].values
    melts = d2['snowmelt'].values
    lons, lats = np.meshgrid(lon, lat)
    return d2,lons,lats

def plot_eof_maps(eofs, eof_num,lons, lats):
    ax = plt.subplot(2, 2, eof_num + 1, projection=ccrs.Stereographic())
    data_crs = ccrs.RotatedPole(pole_longitude=snowmelt_data.rotated_pole.grid_north_pole_longitude,
                                pole_latitude=snowmelt_data.rotated_pole.grid_north_pole_latitude)

    # plot data
    clevs = np.linspace(np.min(eofs[eof_num,0,:,:].squeeze()), np.max(eofs[eof_num,0,:,:].squeeze()), 21)
    result = ax.contourf(lons, lats, eofs[eof_num,0,:,:].squeeze(), clevs, cmap=plt.cm.RdBu_r, transform = data_crs)
    ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '10m', edgecolor='k'))
    cb= plt.colorbar(result, orientation='vertical', extend='both', fraction=0.046, pad=0.04)
    cb.set_label('EOF', fontsize=12)
    variance=round(((varfrac.values[eof_num])*100),2)
    plt.title('EOF ' + str(eof_num+1)+ ', '+str(variance)+'% of total variance ', fontsize=12)
    
    return 

def plot_pc_amplittude(pcs,data_file,pc_num):
    ax=plt.subplot(2,2,pc_num+1)
    days = data_file.time.values
    plt.plot(days, pcs[:,pc_num], linewidth=2)
    plt.axhline(0, color='k')
    plt.xlabel('Day')
    plt.ylabel('PC Amplitude')
    plt.xticks(rotation=45)
    plt.ylim(np.min(pcs.squeeze()), np.max(pcs.squeeze()))
    variance=round(((varfrac.values[pc_num])*100),2)
    plt.title('EOF ' + str(pc_num+1)+ ', '+str(variance)+'% of total variance ', fontsize=12)
```

# Biweekly trends

## EOF and plot BIWEEKLY ALL SEASONS

```python
biweekly_data,lons,lats=resample_bounds(gviis_area,'SMS')
```

```python
coslat = np.cos(np.deg2rad(biweekly_data.coords['rlat'].values))
wgts = np.sqrt(coslat)[..., np.newaxis]
solver = Eof(biweekly_data.snowmelt, weights=wgts)

eof1 = solver.eofs()
pc1  = solver.pcs()
varfrac = solver.varianceFraction()
lambdas = solver.eigenvalues()


plt.figure(figsize=(10, 10))
plt.subplots_adjust(hspace=0.2)
plt.suptitle("Snowmelt EOF Map, 2016 to 2020 biweekly, all seasons", fontsize=20)

for i in range(0,4):
    plot_eof_maps(eof1,i,lons,lats)
    
if saveIt:
    plt.savefig('2016 to 2020 snowmelt EOF biweekly all seasons.png')   
```

```python
plt.figure(figsize=(10, 10))
plt.subplots_adjust(hspace=0.2)
plt.suptitle("Snowmelt PC Amplitude, 2016 to 2020 biweekly, all seasons", fontsize=20)

for i in range(0,4):
    plot_pc_amplittude(pc1,biweekly_data,i)
    
if saveIt:
    plt.savefig('2016 to 2020 snowmelt EOF PC Amplitude biweekly all seasons.png')    
```
