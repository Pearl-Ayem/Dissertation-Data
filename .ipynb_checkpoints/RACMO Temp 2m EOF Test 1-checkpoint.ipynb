```python
from pathlib import Path
import numpy as np
import datetime
import netCDF4
import matplotlib.pyplot as plt
import pyproj
import xarray as xr
import cartopy.feature as cfe
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.colors as colors
import pandas as pd
from pyproj import Transformer
from pyproj import CRS
from numpy import ma
from sklearn import preprocessing
from eofs.xarray import Eof
# from mpl_toolkits.basemap import Basemap
```

```python
filename="t2m_yearlyA_XPEN055_ERAINx_RACMO2.3p2_1979_2018.nc"
this_dir=Path.cwd()
data_dir=this_dir.parent/Path("Data")
files = data_dir.rglob(filename)
for i in files:
    data = i
```

```python
temp2m  = xr.open_dataset(data)
temp2m
```

## Decimal Coordinates of boundary and AWS locations

```python
## Points to bound GVIIS area:
ul_lat= -68.8057057
ul_lon= -70.4385094
ur_lat= -68.9444216
ur_lon= -65.1641375
ll_lat= -73.5935193
ll_lon= -78.5115361
lr_lat= -74.0007447
lr_lon= -67.0772934


#SPRI AWS STATION COORDINATE
AWSlon = -67.69971
AWSlat = -71.14409

#BAS FOSSIL BLUFF AWS STATION COORDINATE
BAS_FB_lon = -68
BAS_FB_lat = -71.329
```

## Define Transformation

```python
data_crs=CRS('-m 57.295779506 +proj=ob_tran +o_proj=latlon +o_lat_p=-180.0 +lon_0=30.0')
from_lat_lon_crs = CRS.from_proj4("+proj=latlon")
transproj = Transformer.from_crs(from_lat_lon_crs, data_crs)
```

## Rotate the points

```python
rotateds= transproj.transform((ll_lon,ur_lon), (ul_lat,lr_lat))
print('data slicing boundary = ',rotateds)
lats_r=rotateds[1]
lons_r=rotateds[0]

##Boundary points rotated
rotated_pts=transproj.transform((ul_lon,ur_lon,ll_lon,lr_lon), (ul_lat,ur_lat,ll_lat,lr_lat))
gviis_lons=rotated_pts[0]
print('GVIIS boundary lons =',gviis_lons)
gviis_lats=rotated_pts[1]
print('GVIIS boundary lons =',gviis_lats)

##SPRI AWS STATION ROTATED
aws_rotated=transproj.transform(AWSlon,AWSlat)
print('AWS rotated = ',aws_rotated)


##BAS FOSSIL BLUFF AWS STATION ROTATED
bas_fb_rotated=transproj.transform(BAS_FB_lon,BAS_FB_lat)
print('BAS Fossil Bluff rotated = ',bas_fb_rotated)
```

# Plot Data with markers for rotated GVIIS boundary and AWS stations

```python
temp=temp2m.t2m.squeeze()
data2= temp.isel(time=0)
rlon = temp2m.rlon.values
rlat = temp2m.rlat.values


plt.figure(figsize=(10,10))

ax = plt.subplot( projection=ccrs.Stereographic(central_longitude=-67.69971, central_latitude=-71.14409))
ax.set_extent([-79,-65,-75,-66])

data_crs = ccrs.RotatedPole(pole_longitude=temp2m.rotated_pole.grid_north_pole_longitude,
                            pole_latitude=temp2m.rotated_pole.grid_north_pole_latitude)

# plot data
result = ax.contourf(rlon, rlat, data2, extend='both', cmap='Blues', transform=data_crs)
ax.plot(gviis_lons[0], gviis_lats[0], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[1], gviis_lats[1], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[2], gviis_lats[2], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[3], gviis_lats[3], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(aws_rotated[0], aws_rotated[1], markersize=7, marker='^', color='yellow',transform=data_crs,label='SPRI AWS')
ax.plot(bas_fb_rotated[0], bas_fb_rotated[1], markersize=7, marker='^', color='orange',transform=data_crs,label='BAS AWS')

plt.legend()

ax.coastlines(resolution='50m')
ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '50m', edgecolor='k'))
plt.colorbar(result, orientation='horizontal', label='K', extend='both', fraction=0.046, pad=0.04)
plt.show()
```

## Subset Data using slicing boundary

```python
gviis_area = temp2m.sel(rlat=slice(lats_r[0],lats_r[1]), rlon=slice(lons_r[0],lons_r[1]))
```

# Raw Plot of the Subset Area only to check if it is GVIIS

```python
gviis_area.t2m[1].plot()
```

## Subset Time

```python
sub1979_1999=gviis_area.sel(time=slice('1979','1999'))
```

```python
lat = sub1979_1999['rlat'].values
lon = sub1979_1999['rlon'].values
temps = sub1979_1999['t2m'].values
lons, lats = np.meshgrid(lon, lat)
nt,height,nlat,nlon=temps.shape
```

## Try direct EOF library and solver

```python
coslat = np.cos(np.deg2rad(sub1979_1999.coords['rlat'].values))
wgts = np.sqrt(coslat)[..., np.newaxis]
solver = Eof(sub1979_1999.t2m, weights=wgts)
```

```python
eof1 = solver.eofs(neofs=10)
pc1  = solver.pcs(npcs=10, pcscaling=0)
varfrac = solver.varianceFraction()
lambdas = solver.eigenvalues()
```

```python
for i in range(0,3):
    fig= plt.figure(figsize=(10,10))
    ax = plt.subplot(projection=ccrs.Stereographic())
    data_crs = ccrs.RotatedPole(pole_longitude=temp2m.rotated_pole.grid_north_pole_longitude,
                                pole_latitude=temp2m.rotated_pole.grid_north_pole_latitude)

    # plot data
    clevs = np.linspace(np.min(eof1[i,0,:,:].squeeze()), np.max(eof1[i,0,:,:].squeeze()), 21)
    result = ax.contourf(lats, lons, eof1[i,0,:,:].squeeze(), clevs, cmap=plt.cm.RdBu_r, transform = data_crs)
    ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '50m', edgecolor='k'))
    cb= plt.colorbar(result, orientation='vertical', extend='both', fraction=0.046, pad=0.04)
    cb.set_label('EOF', fontsize=12)
    plt.title('EOF ' + str(i+1), fontsize=16)
```

```python
for i in range(0,3):
    fig= plt.figure(figsize=(5,5))
    days = np.linspace(1979,1999,nt)
    plt.plot(days, pc1[:,i], linewidth=2)
    plt.axhline(0, color='k')
    plt.xlabel('Year')
    plt.ylabel('PC Amplitude')
    plt.xticks(rotation=45)
    plt.ylim(np.min(pc1.squeeze()), np.max(pc1.squeeze()))
    plt.title('EOF ' + str(i+1), fontsize=16)
```

## Check Variances explained by leading EOFs

```python
plt.figure(figsize=(11,6))
eof_num = range(1, 20)
plt.plot(eof_num, varfrac[0:19], linewidth=2)
plt.plot(eof_num, varfrac[0:19], linestyle='None', marker="o", color='r', markersize=8)
plt.xticks(range(1, 20))
plt.title('Fraction of the total variance represented by each EOF')
plt.xlabel('EOF #')
plt.ylabel('Variance Fraction')
```
