```python
AWSlon = -71.14409
AWSlat = -67.69971
```

```python
# Set the spatial extent to cover the CONUS (Continental United States)
extent = [-78,-66,-74,-69]
central_lon = -67.69971
central_lat = -71.14409

# Albers equal area is a common CRS used to make maps of the United States
f, ax = plt.subplots(figsize=(12, 6),
                     subplot_kw={'projection': ccrs.Stereographic(central_lon, central_lat)})
ax.coastlines()

# Plot the selected location
ax.plot(AWSlat,AWSlon,
        '*', 
        transform=ccrs.PlateCarree(),
        color="purple", 
        markersize=10)

ax.set_extent(extent)

# Adds continent boundaries to the map
ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '50m', edgecolor='k'))

ax.gridlines()
plt.show()
```

```python
yearly_summary = ds.t2m.groupby('time.year').mean('time', skipna=True)
```

```python
yearly_summary.plot(col='year', col_wrap=4, figsize=(20,20))
```

## Made multiple plots for GVIIS extent but formatting is squished 

```python
extent = [-78,-66,-74,-69]
central_lon = -67.69971
central_lat = -71.14409

map_proj = ccrs.AlbersEqualArea(central_longitude=central_lon,
                                central_latitude=central_lat)

aspect = yearly_summary.shape[3] / yearly_summary.shape[2]


p = yearly_summary.plot(transform=ccrs.PlateCarree(),
                        col='year', 
                        col_wrap=2,
                        aspect=aspect,
                        subplot_kws={'projection': map_proj})  # the plot's projection

# Add the coastlines to each axis object and set extent
for ax in p.axes.flat:
    ax.coastlines()
    ax.set_extent(extent)
    ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '50m', edgecolor='k'))
#     ax.plot(AWSlat,AWSlon,
#         '*', 
#         transform=ccrs.PlateCarree(),
#         color="purple", 
#         markersize=10)
```
