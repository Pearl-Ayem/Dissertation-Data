```python
from pathlib import Path
import pandas as pd
import numpy as np
from datetime import datetime, date, time, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.offline as py
```

```python
def read_file_paths(filename):
    this_dir=Path.cwd()
    data_dir=this_dir.parent/Path("Data")
    files = data_dir.rglob(filename)
    for i in files:
        data = i
    return data
```

```python
filename="Fossil_Bluff met data_1979-2020.txt"
colnames=['year','month','day','hour','win_dir','win_sp','slp','temp','dewp','hum','stn_lp']
df = pd.read_csv(str(read_file_paths(filename)), sep=" ", names=colnames, header=None, skiprows =[0,1],
                 parse_dates=[[0,1,2,3]])
```

```python
# time_mask = (df['year_month_day_hour'].dt.year == 1999) & (df['year_month_day_hour'].dt.month == 3)
# df[time_mask]
```

```python
df=df.set_index('year_month_day_hour')
df.head()
```

```python
df.plot(subplots=True, title="Raw Plots of all data",layout=(4, 2), figsize=(15, 15), sharex=True, sharey=False, legend=True)
```

## Resampling 

```python
df_trends = df.resample("AS").agg(['mean', 'min', 'max'])
```

```python
df_trends.head()
```

```python
# pd.options.plotting.backend = "plotly"
fig = df_trends.temp.plot(title="Temperature Trends Fossil Bluff")
```

## Early vs Late Histograms

```python
early = df[datetime(1979,1,1):datetime(2000,1,1)]
late=df[datetime(2000,1,1):datetime(2020,12,31)]
```

```python
## first plot the averages for the first 10 years in blue
early.temp.plot.hist(bins = 20, histtype = 'bar', alpha = 0.5, 
                   rwidth = 0.5, label = "1979-1999")

## then plot the averages for the last 10 years in orange
late.temp.plot.hist(bins = 20, histtype = 'bar', alpha = 0.8, 
                  rwidth = 0.5, label = "2000-2020")

## then add a legend
plt.legend()
```
