```python
from pathlib import Path
import numpy as np
import datetime
import netCDF4
import matplotlib.pyplot as plt
import pyproj
import xarray as xr
import cartopy.feature as cfe
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.colors as colors
import pandas as pd
```

```python
filename="t2m_yearlyA_XPEN055_ERAINx_RACMO2.3p2_1979_2018.nc"
this_dir=Path.cwd()
data_dir=this_dir.parent/Path("Data")
files = data_dir.rglob(filename)
for i in files:
    data = i
```

```python
temp2m  = xr.open_dataset(data)
temp2m
```

## coordinates of the AWS
- Latitude -71.14409
- Longitude -67.69971


```python
AWSlon = -71.14409
AWSlat = -67.69971
```

```python
temp2m.t2m
```

```python
temp=temp2m.t2m.squeeze()
```

```python
temp2m.rotated_pole
```

```python
rad2deg = 180./np.pi
p = pyproj.Proj('+ellps=WGS84 +proj=ob_tran +o_proj=latlon +o_lat_p=-180.0 +o_lon_p=-150.0 +lon_0=180.0')
rlon = temp2m.rlon.values
rlat = temp2m.rlat.values
x1,y1 = np.meshgrid(rlon, rlat)
lon, lat = p(x1, y1) 
lon, lat = lon*rad2deg, lat*rad2deg 
```

```python
data2= temp.isel(time=0)
data2
```

```python
time_bounds = temp2m.time_bnds.isel(time=0)
time_bounds.values
```

```python
tb_values = [pd.Timestamp(tbv).to_pydatetime() for tbv in time_bounds.values]
tb_values
```

```python
time_fmt = '%Y-%m-%d %H:%M:%S'
time_bounds_str = 'From {} to {}'.format(tb_values[0].strftime(time_fmt), tb_values[1].strftime(time_fmt))
time_bounds_str
```

```python
plt.figure(figsize=(10,10))

ax = plt.subplot( projection=ccrs.Stereographic(central_longitude=-67.69971, central_latitude=-71.14409) )
ax.set_extent([-78,-66,-74,-66], ccrs.PlateCarree())

# set up a Cartopy coordinate reference system for the data (rotated pole as described above)
data_crs = ccrs.RotatedPole(pole_longitude=temp2m.rotated_pole.grid_north_pole_longitude,
                            pole_latitude=temp2m.rotated_pole.grid_north_pole_latitude)

# norm = colors.BoundaryNorm(boundaries=levels, ncolors=256)

# plot data
result = ax.contourf(rlon, rlat, data2, extend='both', cmap='Blues', transform=data_crs)

ax.coastlines(resolution='50m')

# oh, and add the ice shelves too (see available features and resolutions at https://www.naturalearthdata.com/features/)
ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '50m', edgecolor='k'))

plt.colorbar(result, orientation='horizontal', label='K', extend='both', fraction=0.046, pad=0.04)

ax.set_title(data2.long_name+'\n'+time_bounds_str+'\n', size='xx-large')

print('')
```

## Plot with real lat/lon with AWS position marked

```python
plt.figure(figsize=(10,10))

projection=ccrs.Stereographic(central_longitude=-67.69971, central_latitude=-71.14409)
plot_coords = projection.transform_points(ccrs.PlateCarree(), lon, lat)
x, y = plot_coords[:,:,0], plot_coords[:,:,1]

ax = plt.subplot(projection=projection)

ax.set_extent([-78,-66,-74,-66], ccrs.PlateCarree())

# set up a Cartopy coordinate reference system for the data (rotated pole as described above)
data_crs = ccrs.RotatedPole(pole_longitude=temp2m.rotated_pole.grid_north_pole_longitude,
                            pole_latitude=temp2m.rotated_pole.grid_north_pole_latitude)


# plot data 
result = ax.contourf(x, y, data2, extend='both', cmap='Blues',transform=projection)
ax.plot(AWSlon, AWSlat, transform=projection,markersize=5, marker='o', color='red')

ax.coastlines(resolution='50m')

# oh, and add the ice shelves too (see available features and resolutions at https://www.naturalearthdata.com/features/)
ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '50m', edgecolor='k'))

plt.colorbar(result, orientation='horizontal', label='K', extend='both', fraction=0.046, pad=0.04)

ax.set_title(data2.long_name+'\n'+time_bounds_str+'\n', size='xx-large')

print('')
```
