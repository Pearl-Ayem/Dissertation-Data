## Decimal Coordinates of boundary and AWS locations

```python
## Points to bound GVIIS area:
ul_lat= -68.8057057
ul_lon= -70.4385094
ur_lat= -68.9444216
ur_lon= -65.1641375
ll_lat= -73.5935193
ll_lon= -78.5115361
lr_lat= -74.0007447
lr_lon= -67.0772934


#SPRI AWS STATION COORDINATE
AWSlon = -67.69971
AWSlat = -71.14409

#BAS FOSSIL BLUFF AWS STATION COORDINATE
BAS_FB_lon = -68
BAS_FB_lat = -71.329
```

## Define Transformation

```python
snowmelt_data.rotated_pole
```

```python
data_crs=CRS('-m 57.295779506 +proj=ob_tran +o_proj=latlon +o_lat_p=-180.0 +lon_0=30.0')
from_lat_lon_crs = CRS.from_proj4("+proj=latlon")
transproj = Transformer.from_crs(from_lat_lon_crs, data_crs)
```

## Rotate the points

```python
rotateds= transproj.transform((ll_lon,ur_lon), (ul_lat,lr_lat))
print('data slicing boundary = ',rotateds)
lats_r=rotateds[1]
lons_r=rotateds[0]

##Boundary points rotated
rotated_pts=transproj.transform((ul_lon,ur_lon,ll_lon,lr_lon), (ul_lat,ur_lat,ll_lat,lr_lat))
gviis_lons=rotated_pts[0]
print('GVIIS boundary lons =',gviis_lons)
gviis_lats=rotated_pts[1]
print('GVIIS boundary lons =',gviis_lats)

##SPRI AWS STATION ROTATED
aws_rotated=transproj.transform(AWSlon,AWSlat)
print('AWS rotated = ',aws_rotated)


##BAS FOSSIL BLUFF AWS STATION ROTATED
bas_fb_rotated=transproj.transform(BAS_FB_lon,BAS_FB_lat)
print('BAS Fossil Bluff rotated = ',bas_fb_rotated)
```

```python
melt=snowmelt_data.snowmelt.squeeze()
data2= melt.isel(time=0)
rlon = snowmelt_data.rlon.values
rlat = snowmelt_data.rlat.values


plt.figure(figsize=(10,10))

ax = plt.subplot( projection=ccrs.Stereographic(central_longitude=-67.69971, central_latitude=-71.14409))
ax.set_extent([-79,-65,-75,-66])

data_crs = ccrs.RotatedPole(pole_longitude=snowmelt_data.rotated_pole.grid_north_pole_longitude,
                            pole_latitude=snowmelt_data.rotated_pole.grid_north_pole_latitude)

# plot data
result = ax.contourf(rlon, rlat, data2, extend='both', cmap='Blues', transform=data_crs)
poly_coords = Polygon(((gviis_lons[0], gviis_lats[0]),
                       (gviis_lons[1], gviis_lats[1]),
                       (gviis_lons[3], gviis_lats[3]),
                       (gviis_lons[2], gviis_lats[2])))


# ax.add_patch(Polygon(poly_coords, color='forestgreen', alpha=0.5))
ax.plot(gviis_lons[0], gviis_lats[0], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[1], gviis_lats[1], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[2], gviis_lats[2], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(gviis_lons[3], gviis_lats[3], markersize=7, marker='o', color='red',transform=data_crs)
ax.plot(aws_rotated[0], aws_rotated[1], markersize=7, marker='^', color='green',transform=data_crs,label='SPRI AWS')
ax.plot(bas_fb_rotated[0], bas_fb_rotated[1], markersize=7, marker='^', color='orange',transform=data_crs,label='BAS AWS')

plt.legend()

ax.coastlines(resolution='10m')
ax.add_feature(cfe.NaturalEarthFeature('physical', 'antarctic_ice_shelves_lines', '10m', edgecolor='k'))
plt.title("Snowmelt Flux on the GVIIS")
plt.colorbar(result, orientation='vertical', label='kg $m^{-2}$ $s^{-1}$', extend='both', fraction=0.046, pad=0.04)
ax.add_geometries([poly_coords], crs=data_crs, facecolor="none",edgecolor='red', alpha=0.8,linewidth=5)

plt.show()
```

## Subset Data using slicing boundary

```python
gviis_area = snowmelt_data.sel(rlat=slice(lats_r[0],lats_r[1]), rlon=slice(lons_r[0],lons_r[1]))
```

```python
# create a function with:
# input: list of months,year,data
# output: number of days in year y, that all months in m_list saw snowmelt

def snowmelt_days_month_list(m_list,y,data):                          
    output_per_year=[]
    for m in m_list:
        x=snowmelt_days(m,y,data)
        output_per_year.append(x.time.shape[0])
    return output_per_year

# create a function with:
# input: month,year,data
# output: dataarray of the month m in year y of all the points on days when melt occured 

def snowmelt_days(m,y,data):                                                 #(types: int,int,xarray dataframe)
    daily_max=data.snowmelt.resample(time='1D').max()                        #find daily max snowmelt
    month_data=daily_max.sel(time=daily_max.time.dt.month.isin([m]) )        # filter to only include the data from the input month
    year_month_data=month_data.sel(time=month_data.time.dt.year.isin([y]))   # filter to only include the data from the input month
    melt_days=year_month_data.where(lambda x: x >0, drop=True)               # filter to only include days with snowmelt
    return melt_days

```

```python
months=[10,11,2,3,4]
years=[2016,2017,2018,2019,2020]
all_years=[]
for y in years:
    all_years.append(snowmelt_days_month_list(months,y,gviis_area))
all_years 
```

```python
melt_days = pd.DataFrame(columns=['year','month','num_melt_days'])
for i in range(len(years)):
    for j in range(len(months)):
        df_new_row = pd.DataFrame({'year': years[i],'month':months[j],'day':1,'num_melt_days':all_years[i][j]},index=[0])
        melt_days = pd.concat([melt_days,df_new_row])        
```

```python
melt_days['date']=pd.to_datetime(melt_days[['year', 'month','day']])
del melt_days['day']
melt_days=melt_days.set_index('date')
monthly=melt_days.groupby(['month'],as_index=False).num_melt_days.mean()
```

```python
df_pivot = pd.pivot_table(melt_days,
                          values="num_melt_days",
                          index="month",
                          columns="year",
                          aggfunc=np.mean)
ax1=df_pivot.plot(kind='bar',sharex=True,figsize=(8,8),
              title= 'Number of melt days per month on the GVIIS')
monthly.num_melt_days.plot(ax=ax1,color='k')
```
