```python
from pathlib import Path
import numpy as np
import datetime
import netCDF4
import matplotlib.pyplot as plt
import pyproj
import xarray as xr
import cartopy.feature as cfe
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.colors as colors
import pandas as pd
from numpy import ma
from sklearn import preprocessing
```

```python
filename="Wet Station Table 1 clean_file.xlsx"
this_dir=Path.cwd()
p = this_dir.parent
data = [f for f in p.glob('**/Wet Station Table 1 clean_file.xlsx')][0]
data
```

```python
df_raw = pd.read_excel(str(data),
                    header=0,
                    skiprows=[0,2,3],
                    usecols=[0,4,5,8,9,10,11,12,13,14],
                    converters= {'TIMESHEET': pd.to_datetime},
                    na_values=['NAN'],
                    index_col=0)
```

```python
def resample_summer(data,summer_months):
    start_date='2019-11-17 08:45:00'
    mask=data.index.month.isin(summer_months) & (data.index > start_date)
    return data.loc[mask]
```

```python
df=resample_summer(df_raw,[11,12,1,2,3])
biweekly = df.resample('SMS').mean().dropna()
```
