```python
filename="Wet Station Table 1 clean_file.xlsx"
this_dir=Path.cwd()
p = this_dir.parent
data = [f for f in p.glob('**/Wet Station Table 1 clean_file.xlsx')][0]
data
```

```python
df1 = pd.read_excel(str(data),
                    header=0,
                    skiprows=[0,2,3],
                    usecols=[0,4,5,8,9,10,11,12,13,14],
                    converters= {'TIMESHEET': pd.to_datetime},
                    na_values=['NAN'],
                    index_col=0)
```

```python
df1.info()
```

## Mask dates to only include data after Nov, 17 2019

```python
start_date='2019-11-17 08:45:00'
end_date = '2020-04-30 21:15:00'
mask = (df1.index > start_date) & (df1.index < end_date)
df1 = df1.loc[mask]
df1.head()
```

## Covert timezone from UTC to Antarctica

```python
antarctica_time = pytz.timezone('Antarctica/Rothera')
df1.index = df1.index.tz_localize(pytz.utc).tz_convert(antarctica_time)
df1.head()
```

## Subset data to include months of Nov to April

```python
sum_aws= df1[df1.index.month.isin([11,12,1,2,3,4])]
```

## Resample to 3hour averages

```python
sum_aws_3h=sum_aws.resample('3H').mean()
```

## Resample biweekly

```python
biweekly_aws=sum_aws.resample('SMS').mean()
```

```python
filename="rh2m.KNMI-2016.XPEN055.ERA5-3H_RACMO2.3p2.3H.nc"
this_dir=Path.cwd()
data_dir=this_dir.parent/Path("Data")
files = data_dir.rglob(filename)
for i in files:
    data = i
```

```python
rh_data_racmo  = xr.open_dataset(data)
rh_data_racmo
```

## GVIIS Boundary and Data points for FB and Wet AWS

```python
## Points to bound GVIIS area:
ul_lat= -68.8057057
ul_lon= -70.4385094
ur_lat= -68.9444216
ur_lon= -65.1641375
ll_lat= -73.5935193
ll_lon= -78.5115361
lr_lat= -74.0007447
lr_lon= -67.0772934


#SPRI AWS STATION COORDINATE
AWSlon = -67.69971
AWSlat = -71.14409

#BAS FOSSIL BLUFF AWS STATION COORDINATE
BAS_FB_lon = -68
BAS_FB_lat = -71.329
```

## Define Transformation

```python
data_crs=CRS('-m 57.295779506 +proj=ob_tran +o_proj=latlon +o_lat_p=-180.0 +lon_0=30.0')
from_lat_lon_crs = CRS.from_proj4("+proj=latlon")
transproj = Transformer.from_crs(from_lat_lon_crs, data_crs)
```

## Rotate Points

```python
rotateds= transproj.transform((ll_lon,ur_lon), (ul_lat,lr_lat))
print('data slicing boundary = ',rotateds)
lats_r=rotateds[1]
lons_r=rotateds[0]

##Boundary points rotated
rotated_pts=transproj.transform((ul_lon,ur_lon,ll_lon,lr_lon), (ul_lat,ur_lat,ll_lat,lr_lat))
gviis_lons=rotated_pts[0]
print('GVIIS boundary lons =',gviis_lons)
gviis_lats=rotated_pts[1]
print('GVIIS boundary lons =',gviis_lats)

##SPRI AWS STATION ROTATED
aws_rotated=transproj.transform(AWSlon,AWSlat)
print('AWS rotated = ',aws_rotated)


##BAS FOSSIL BLUFF AWS STATION ROTATED
bas_fb_rotated=transproj.transform(BAS_FB_lon,BAS_FB_lat)
print('BAS Fossil Bluff rotated = ',bas_fb_rotated)
```

## Extract data for just Fossil Bluff and Wet AWS locations

```python
racmo_aws_rh = rh_data_racmo.rh2m.sel(rlon=aws_rotated[0], rlat=aws_rotated[1], method='nearest')
racmo_fb_rh=rh_data_racmo.rh2m.sel(rlon=bas_fb_rotated[0], rlat=bas_fb_rotated[1], method='nearest')
```

## Subset Time and convert to pandas

```python
start_date='2019-11-17 08:45:00'
end_date = '2020-04-30 21:15:00'
racmo_fb_rh=racmo_fb_rh.sel(time=slice(start_date,end_date))
racmo_aws_rh=racmo_aws_rh.sel(time=slice(start_date,end_date))
```

```python
racmo_fb_rh = racmo_fb_rh.to_dataframe()
racmo_fb_rh= racmo_fb_rh.reset_index(level=['height'])
racmo_aws_rh = racmo_aws_rh.to_dataframe()
racmo_aws_rh= racmo_aws_rh.reset_index(level=['height'])
```

```python
racmo_fb_rh.index
```

## Convert RACMO timezone from UTC to Rothera as well

```python
antarctica_time = pytz.timezone('Antarctica/Rothera')
racmo_fb_rh.index = racmo_fb_rh.index.tz_localize(pytz.utc).tz_convert(antarctica_time)
racmo_aws_rh.index = racmo_aws_rh.index.tz_localize(pytz.utc).tz_convert(antarctica_time)
racmo_aws_rh.head()
```

```python
racmo_fb_rh.head()
```

```python
racmo_fb_rh['rh2m'] = racmo_fb_rh['rh2m'].apply(lambda x: x*100)
racmo_aws_rh['rh2m'] = racmo_aws_rh['rh2m'].apply(lambda x: x*100)
```

## Resample to 3hour averages

```python
racmo_rh_aws_3h=racmo_aws_rh.resample('3H').mean()
racmo_rh_fb_3h=racmo_fb_rh.resample('3H').mean()
```

## Resample biweekly

```python
racmo_rh_aws_biweekly=racmo_aws_rh.resample('15D').mean()
racmo_rh_fb_biweekly=racmo_fb_rh.resample('15D').mean()
```
