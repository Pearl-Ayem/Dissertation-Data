```python
from pathlib import Path
import pandas as pd
import numpy as np
import datetime
```

```python
data_dir_name ="AWS"
filename="Wet Station Table1.xlsx"
this_dir=Path.cwd()
data_dir=this_dir.parent/Path(data_dir_name)
data=data_dir/Path(filename)
```
